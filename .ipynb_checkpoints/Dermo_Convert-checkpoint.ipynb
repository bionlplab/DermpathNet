```python
import pandas as pd
import os

# Your given paths
image_folder_path = '/Users/stephaniexu/Downloads/image_filtered/Warty dyskeratoma'
webpage_folder_path = '/Users/stephaniexu/Downloads/image_filtered/Webpages_for_image/Warty dyskeratoma'
csv_file_path = '/Users/stephaniexu/Downloads/meta/Warty dyskeratoma/merged_csv_file_Warty_dyskeratoma.csv'

# Read the CSV file
df = pd.read_csv(csv_file_path)
# Ensure the webpage folder exists
os.makedirs(webpage_folder_path, exist_ok=True)

# List all image files in the directory
image_files = os.listdir(image_folder_path)

# Initialize the JavaScript content variable
js_content = ''

# Loop through each image file to generate HTML content and associated JS
for index, row in df.iterrows():
    # 确保 graphic_name 是字符串类型
    graphic_name = str(row['graphic'])
    id_name = str(row['pmcid'])
    matching_files = [img for img in os.listdir(image_folder_path) if graphic_name in img and id_name in img and img.lower().endswith(('.png', '.jpg', '.jpeg', '.gif'))]
        
    if matching_files:
        # 如果找到匹配的图片文件，处理它们...
        img_file = matching_files[0]  # 取第一个匹配项
        base_name = os.path.splitext(img_file)[0]
        webpage_path = os.path.join(webpage_folder_path, base_name + '.html')
        
        # 创建 HTML 文件
        with open(webpage_path, 'w') as f:
            title = str(row['title'])  # 确保标题是字符串类型
            description = str(row['caption'])  # 确保描述是字符串类型
            PMID = str(row['pmid'])
            f.write(format_html)
            f.write('<div class="content-container">')  # 包裹整个内容的新容器，避免与现有的container类名冲突
            f.write('  <div class="figures-container">')  # 定义一个flex容器，用于并排显示图像和文本

            # 图像容器
            f.write('    <div class="image-side">')  # 图片部分
            f.write('      <img src="{}" alt="{}" class="clickable-image">'.format(image_folder_path+'/'+img_file, title))
            f.write('    </div>')  # 关闭image-side div

            # 文本容器
            f.write('    <div class="text-side">')  # 文本部分，将会出现在图片右侧
            f.write('      <h1>{}</h1>'.format(title))
            f.write('      <p>{}</p>'.format(description))
            pubmed_url = f"https://pubmed.ncbi.nlm.nih.gov/{PMID}/"
            f.write('      <p>PMID: <a href="{}" target="_blank">{}</a></p>'.format(pubmed_url, PMID))

            f.write('    </div>')  # 关闭text-side div

            f.write('  </div>')  # 关闭figures-container div
            
            f.write('</div>')  # 关闭最外层的content-container div
            f.write('</body></html>')

```

### Add the Hierachical Data

```python
import yaml

# Load the YAML file to examine its structure
file_path = 'Dermpath research project glossary_Ver 3.yaml'
with open(file_path, 'r') as file:
    hierarchy_data = yaml.safe_load(file)

def get_category_and_type(subcategory, hierarchy):
    for category, subcats in hierarchy.items():
        if subcategory in subcats:
            return category, subcats[subcategory]['Type']
    return None, None
```

```python
webpage_base_path = 'image_filtered/Webpages_for_image'
```

```python
for subcategory in hierarchy_data.values():
    for subcat_name in subcategory:
        subcat_dir_path = os.path.join(webpage_base_path, subcat_name)
        
        if os.path.isdir(subcat_dir_path):
            category, type_ = get_category_and_type(subcat_name, hierarchy_data)
            
            # Prepare the text to be inserted
            text_to_insert = f"<p>Type: {type_}, Category: {category}</p>\n"

            # Add the text to each HTML file in the subcategory directory
            for html_file in os.listdir(subcat_dir_path):
                if html_file.endswith('.html'):
                    html_file_path = os.path.join(subcat_dir_path, html_file)
                    
                    with open(html_file_path, 'r+', encoding='utf-8') as file:
                        content = file.read()
                        position = content.find('</p>', content.find('<p>PMID:')) + 4
                        if position != -1:
                            content = content[:position] + text_to_insert + content[position:]
                            file.seek(0)
                            file.write(content)
                            file.truncate()
```

```python
import os
from bs4 import BeautifulSoup

# 目录路径
directory_path = '/Users/stephaniexu/Desktop/Dermo HTML/image_filtered/Webpages_for_image/Eccrine Poroma'

# 遍历目录中的所有 HTML 文件
for filename in os.listdir(directory_path):
    if filename.endswith('.html'):  # 确保仅处理 HTML 文件
        file_path = os.path.join(directory_path, filename)
        
        # 读取 HTML 文件内容
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        
        # 使用 BeautifulSoup 解析 HTML
        soup = BeautifulSoup(content, 'html.parser')
        
        # 找到所有 <p> 标签并检查重复项
        paragraphs = soup.find_all('p')
        seen = set()  # 用来跟踪已经看到的 <p> 文本
        for p in paragraphs:
            text = p.get_text(strip=True)
            if text.startswith('Type:') and text in seen:
                p.decompose()  # 如果这段文本已经出现过一次，删除这个 <p> 标签
            else:
                seen.add(text)  # 记录这段文本
        
        # 将修改后的 HTML 写回文件
        with open(file_path, 'w', encoding='utf-8') as file:
            file.write(str(soup))

print("All duplicate paragraphs have been removed from the HTML files.")
```

### change image source to html from NIH

```python
import re

# Define the file path
file_path = 'browser_info.js'

# Read the content of the file
with open(file_path, 'r', encoding='utf-8') as file:
    content = file.read()

# Define the regex pattern to match the img tag structure
pattern = re.compile(r'(<img src="image_filtered/.+?/PMC(\d+)-(.+?)\.jpg" alt=".+?" style="width:100%;" /></a>)')

# Function to replace the matched pattern
def replace_pattern(match):
    pmc_id = match.group(2)
    img_id = match.group(3)
    new_img_src = f'https://www.ncbi.nlm.nih.gov/pmc/articles/PMC{pmc_id}/bin/{img_id}.jpg'
    replaced = f'<img src="{new_img_src}" /></a>'
    return replaced

# Replace all matches in the content
new_content = re.sub(pattern, replace_pattern, content)

# Save the modified content back to the file
with open(file_path, 'w', encoding='utf-8') as file:
    file.write(new_content)

print("Replacement complete.")

```

### Run for each separate page

```python
import os
import re

# 定义包含HTML文件的根目录
root_dir = 'image_filtered/Webpages_for_image/'

# 定义匹配 <img src> 结构的正则表达式模式
pattern = re.compile(r'<img src="\.\./\.\./(.+?)/PMC(\d+)-(.+?)\.jpg" (.+?)>')

# 替换匹配模式的函数
def replace_pattern(match):
    folder = match.group(1)  # 捕获文件夹名称
    pmc_id = match.group(2)  # 捕获PMC编号
    img_id = match.group(3)  # 捕获图片ID
    rest_of_tag = match.group(4)  # 捕获标签的其余部分
    new_img_src = f'https://www.ncbi.nlm.nih.gov/pmc/articles/PMC{pmc_id}/bin/{img_id}.jpg'
    replaced = f'<img src="{new_img_src}" {rest_of_tag}>'
    return replaced

# 遍历目录并处理每个HTML文件
for subdir, dirs, files in os.walk(root_dir):
    for file in files:
        if file.endswith('.html'):
            file_path = os.path.join(subdir, file)
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
            
            # 替换内容中的所有匹配项
            new_content = re.sub(pattern, replace_pattern, content)
            
            # 将修改后的内容保存回文件
            with open(file_path, 'w', encoding='utf-8') as f:
                f.write(new_content)

print("Replacement complete.")
```
